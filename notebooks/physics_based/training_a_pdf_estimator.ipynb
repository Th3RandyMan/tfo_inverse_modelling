```python
import torch
import torch.nn as nn

class PMFEstimatorNet(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(PMFEstimatorNet, self).__init__()
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.model = nn.Sequential(
            # nn.Linear(self.input_dim, self.hidden_dim),
            nn.Linear(self.input_dim, self.output_dim),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.model(x)
        x = x / torch.sum(x, dim=1, keepdim=True)
        return x

input_train = torch.randn(200, 5)
input_val = torch.randn(100, 5)
labels_train = torch.abs(torch.randn(200, 3))
labels_val = torch.abs(torch.randn(100, 3))
# Normalize labels to add upto 1
labels_train = labels_train / torch.sum(labels_train, dim=1, keepdim=True)
labels_val = labels_val / torch.sum(labels_val, dim=1, keepdim=True)
model = PMFEstimatorNet(5, 5, 3)

output = model(input_train)
print(output.size())
```

```python
# Run a training loop with MSE LOss
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
model = model.train() 
for i in range(10000):
    optimizer.zero_grad()
    output = model(input_train)
    loss = criterion(output, labels_train)
    loss.backward()
    optimizer.step()
    if i % 100 == 0:
        print("Loss: ", loss.item())

# Run a validation loop
output_val = model(input_val)
model = model.eval()
with torch.no_grad():
    loss_val = criterion(output_val, labels_val)
print("Validation Loss: ", loss_val.item())
```
